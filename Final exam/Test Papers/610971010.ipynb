下面的程式碼`A`中有幾個標註部分(如`@__1__@`)，請從下方的選項中選擇適合的選項並填入標註處(不會重複)，讓程式能輸出結果`B` :

* 程式碼`A`

```python
list1 = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]

s = @__1__@

for i in range(@__2__@(list1)):
    s += list1[@__3__@]
    
print('Sum of list1 is :', @__4__@)
```

* 程式碼填空選項

1. `0`
2. `2046`
3. `s`
4. `i`
5. `list1`
6. `max`
7. `len`

* 程式輸出結果

```
Sum of list1 is : 2047
```

````python
# ↓ 將適合的選項放在此區塊中的標註處 ↓

list1 = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]

s = @__1__@

for i in range(@__2__@(list1)):
    s += list1[@__3__@]
    
print('Sum of list1 is :', @__4__@)
````
