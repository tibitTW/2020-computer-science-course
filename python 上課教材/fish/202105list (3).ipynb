# 串列list 基本定義


為了讓大量資料的儲存更方便、更有效率，一般會藉助一些利器來處理這些資料，Python 提供優質的利器：串列list，有如一個有組織的置物櫃一般。

串列 ( 又稱為「清單」或「列表」)，與其他語言的「陣列 (Array)」相同，
是一個用來儲存"多個變數"很好的方法。

功能: 與變數相類似，是提供儲存資料的記憶體空間。
每一個串列擁有一個名稱，做為識別該串列的標誌；串列中每一個資料稱為「***元素***」，每一個串列元素相當於一個變數，如此就可輕易儲存大量的資料儲存空間。要存取串列中特定元素，是以元素在***串列中的位置***做為***索引***，即可存取串列元素。

串列有一維串列、二維串列，以及多維串列。若串列沒有指明其維度時，則預設為一維串列（one dimension list）。

串列的使用方式是將元素置於中括號 ([]) 中，元素之間以逗號分隔，語法為：

串列名稱＝[元素1,元素2,元素3......]

串列的長度其實就是串列中有多少個元素。若要計算某一串列的長度，可使用len函式，如:

  fruitsList= ['Banana', 'Apple', 'Grape', 'Pineapple', 'Orange']
  len(fruitsList)

5

上述敘述表示有一fruitsList串列，利用len函式計算此串列的長度為5，計表示它有5個元素。

取得元素值的方法是將索引值置於中括號內，注意索引值是從 0 開始計數：第一個元素值索引值為 0，第二個元素值索引值為 1，依此類推。索引值不可超出串列的範圍，否則執行時會產生錯誤。

```python
fruit2=['apple','banana','orange']
print(fruit1[0])#apple
print(fruit1[1])#banana
print(fruit1[3])#error 索引值超過範圍, IndexError: list index out of range
```

索引值可以是***負值***，表示由串列的最後向前取出，

***「-1」表示最後一個元素***，「-2」表示倒數第二個元素，依此類推。

同理，負數索引值不可超出串列的範圍，否則執行時會產生錯誤。例如：

```python
fruit3=['apple','banana','orange']
print(fruit1[-1])#orange
print(fruit1[-2])#banana
print(fruit1[-4])#error 索引值超過範圍, IndexError: list index out of range
```

# 串列list排序

```python
score=[89,45,78,90,67,89] #數學平時考試成績
fruits=['apple','banana', 'orange','kiwi'] #店內賣的水果種類
score01=['Mary',89,45,78,90,67,89] #Mary數學平時考試成績
score.sort()  #數字排序,由小至大
fruits.sort()  #字串排序,由小至大
print(score)
print(fruits)
score.sort(reverse=True) #由大到小
print(score)
```

```python
#串列排序 sort(), 串列反轉 reverse()
fruits=['papaya','apple','banana', 'orange'] #店內賣的水果種類
fruits.sort()#sort()排序
print(fruits)  
fruits.reverse()#reverse()反轉
print(fruits)
score3=[3,25,6,2,1,78,89,-3] #數字陣列
score3.sort()#sort()排序
print(score3)
score3.reverse()#reverse()反轉
print(score3)
```

```python
#sort()的進階說明01
scores = [('D', 50), ('B', 30), ('E', 20), ('A', 10), ('C', 70)]
scores.sort()
print(scores)
```

```python
#sort()的進階說明01
scores = [('D', 50), ('B', 30), ('E', 20), ('A', 10), ('C', 70)]
scores.sort(key=lambda scores: scores[1]) #用lambda指定排序的資料
print(scores)
```

```python
#sort()的進階說明03
#如果每一筆資料中有多項成績，也可以利用Lamda函式把成績加總，再根據加總的結果排序
scores = [('D', 50, 80), ('B', 30, 90), ('E', 20, 70), ('A', 90, 60), ('C', 70, 60)]
scores.sort(key=lambda scores: scores[1] + scores[2]) 
print(scores)
```

```python
#sort()的進階說明03
#內建函式sorted()。除了List資料組的sort()方法之外，Python還有一個內建函式sorted()也可以進行排序。
#差別是內建函式sorted()它會產生一個新的資料組，不會變更原來資料組的資料順序。
scores = [50, 30, 90, 10, 70]
sorted_scores = sorted(scores)
print(scores)   # 顯示原來的資料
print(sorted_scores)   # 顯示排序後的資料
```

# 串列list增減操作
TextBook p3-15

```python
#append()和insert()方法
fruits=['apple','banana', 'orange'] #店內賣的水果種類
# append() 會將新加入的值放在最後一個
fruits.append('watermelon')
print(fruits)
# insert()中前面的值是所要插入的索引位置
pets = ['cat','dog','bird']
pets.insert(1,'snake') # insert(i, x)，將x元素加入於串列的i位置
print(pets)
```

```python
# remove()和pop()方法
# remove() 做資料的刪除
pets = ['cat', 'snake', 'dog', 'bird']
pets.remove('snake') #remove(x)，直接刪除串列中第一個出現x的元素。
print (pets)
# pop() 做資料的刪除
pets = ['cat', 'snake', 'dog', 'bird']
pets.pop(2) # pop(i)，表示將i 索引的元素值刪除, pop(i) 方法中的i，也可以是負的
print (pets)
```
