```python
import numpy as np
import matplotlib.pyplot as plt
```

### $f(x) = x^2 + 2x + 1$

```python
x = np.arange(-3, 3, 0.1)
y = x ** 2 + 2 * x + 1

plt.xlim(-10, 10)
plt.ylim(-5, 15)
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(np.pi * (-2), np.pi * 2, 0.1)
y1 = np.sin(x * 2) * 5
y2 = np.cos(x)

plt.plot(x, y1, 'r')
plt.plot(x, y2)
```
