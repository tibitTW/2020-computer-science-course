`O` `X` 雙方輪流在3x3的九宮格上下子，若有一方連成一條線，則該玩家就獲勝

> 使用者輸入一串只包含`X`、`O`及` `的字串(大寫英文字母及空白)，代表一局井字遊戲的結果。每個字元分別對應九宮格的每一個格子，順序從九宮格的左上角到右下角 :

```
 1 │ 2 │ 3
───┼───┼───
 4 │ 5 │ 6
───┼───┼───
 7 │ 8 │ 9
```

> 因此若輸入字串為`OXX OXOOX`，則代表以下，`X`方獲勝的一局井字遊戲 :

```
 O │ X │ X
───┼───┼───
   │ O │ X
───┼───┼───
 O │ O │ X
```

* **輸出說明**

> 判斷此局井字遊戲的結果為誰獲勝，如果`X`方則輸出`X`，`O`方則輸出`O`，雙方平手則輸出`DRAW`

* **example 1**

    * Input
    
    > `XOOOXXOXO`
    
    * Output
    
    > `DRAW`
    
    * Explanation
    
        這局的棋盤狀態如下 :

        ```
         X │ O │ O
        ───┼───┼───
         O │ X │ X
        ───┼───┼───
         O │ X │ O
        ```
